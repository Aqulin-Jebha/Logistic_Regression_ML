```python
cols=['customerID']
df=data.drop(cols,axis="columns")
df.head()
```

```python
pd.to_numeric(df['TotalCharges'],errors='coerce')
df[pd.to_numeric(df['TotalCharges'],errors='coerce').isnull()].shape
df=df[df['TotalCharges']!=' ']
```

```python
def print_unique(df):
    for cols in df:
        if df[cols].dtypes=='object':
            print(f'{cols}: {df[cols].unique()}')
```

```python
print_unique(df)
```

```python
df.replace('No phone service','No', inplace=True)
df.replace('No internet service','No',inplace=True)
print_unique(df)
```

```python
colums =['Partner','Dependents','PhoneService','OnlineSecurity','MultipleLines','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']
for i in colums:
    print(i)
    df[i].replace({'Yes': 1,'No' :0}, inplace=True)
```

```python
df.head(10)
```

```python
df['gender'].replace({'Male' :1, 'Female' :0}, inplace=True)
df.head()
```

```python
cols=['Contract', 'PaymentMethod','InternetService']
df1 = pd.get_dummies(data=df,columns=cols)
df1.columns
```

# scaling data

```python
df1.head()
```

```python
cols_to_scale = ['tenure','MonthlyCharges','TotalCharges']
from sklearn.preprocessing import MinMaxScaler
scalar= MinMaxScaler()
df1[cols_to_scale] = scalar.fit_transform(df1[cols_to_scale])
df1[cols_to_scale]
```

```python
df1.columns
```

```python
df1.head(10)
```

```python
df1.info()
```

```python
X= df1.drop('Churn',axis='columns')
y= df1['Churn']
```

```python
from sklearn.model_selection import  train_test_split
X_train,X_test,y_train,y_test =train_test_split(X,y, test_size=0.2, random_state=10)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_predict = model.predict(X_test)
y_predict
```

```python
model.score(X_test,y_test)
```
